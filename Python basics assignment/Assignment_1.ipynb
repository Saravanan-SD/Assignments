10. Why does this expression cause an error? How can you fix it?
'I have eaten ' + 99 + ' burritos.'

The error is we can only concatenate string to string by this expression is trying to concatenate string with a integer.

This can be fixed by type casting the integer value to string.

'I have eaten ' + '99' + ' burritos.'

This will give a correct output:- 'I have eaten 99 burritos.'
