10. How can you tell the difference between break and continue?

(i) break eliminates the execution of the remaining iteration of loop whereas continue terminates only the current iteration of the loop.

(ii) break terminates the loop whereas continue causes the early execution of the next iteration.

(iii) break will resume control of the program to the end of the loop enclosing the break whereas continue will resume the control of the program to next iteration of that loop enclosing 'continue'.
