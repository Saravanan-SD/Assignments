```python
#1. Write a Python program to convert kilometers to miles?
km=int(input('Enter the Kilometer: '))
miles= round((km/1.609),4)
print(f'{km} Kilometer in Miles is {miles} approximately')
```

```python
#2. Write a Python program to convert Celsius to Fahrenheit?

c=int(input('Enter temperture in celcius: '))
f=(c*(9/5))+32
print(f'Temperatur in Fahrenheit is {f}')
```

```python
3. Write a Python program to display calendar?
import calendar

yy=int(input("Enter year: "))
mm=int(input("Enter month: "))

print(calendar.month(yy,mm))
```

```python
#4. Write a Python program to solve quadratic equation?
import cmath
A=float(input("Enter Coeeficient of A: "))
B=float(input("Enter Coeeficient of B: "))
C=float(input("Enter Coeeficient of C: "))
d=(B**2-4*A*C)
if d>0:
    r1=(-B+cmath.sqrt(d))/(2*A)
    r2=(-B-cmath.sqrt(d))/(2*A)
    print(format(r1, '.2f'))
    print(format(r2, '.2f'))
elif d==0:
    r1=(-B)/2*A
    print(format(r1, '.2f'))
    
elif d<0:
    r1=(-B+cmath.sqrt(d))/(2*A)
    r2=(-B-cmath.sqrt(d))/(2*A)
    print(format(r1, '.2f'))
    print(format(r2, '.2f'))
else:
    print("Error")
```
